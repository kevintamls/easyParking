```python
import pandas as pd
import numpy as np
import json
import geopandas as gpd
import requests
from shapely.geometry import Point
```

```python
# read the parking spaces dataset from the government database
parking_df = pd.read_csv('ON_STREET_PARK.csv')
parking_df
```

```python
# old coordinates
x_old_coor = parking_df['X_COOR']
y_old_coor = parking_df['Y_COOR']
```

```python
# convert old coordinates to standardized coordinates
points = []
for i in range(len(parking_df)):
    url = 'http://www.geodetic.gov.hk/transform/v2/?inSys=hkgrid&outSys=wgsgeog&e=' + str(x_old_coor[i]) + '&n=' + str(y_old_coor[i])
    r = requests.get(url)
    data = r.json()
    coor = Point((data['wgsLong'], data['wgsLat']))
    points.append(coor)
```

```python
# plot all parking spaces on the web map
d = {'geometry': points}
gdf = gpd.GeoDataFrame(d, crs = 'EPSG:3857')
geojsonio.display(gdf.to_json())
```
